```python
import pandas as pd
data = pd.read_csv("/content/data for ner.csv",encoding="latin1" )
```

```python
data.head(30)
```

```python
train_data = pd.DataFrame({"sentence_id":data["sentence_id"],"words":data["words"],"labels":data['labels']})
```

```python
train_data
```

```python
from simpletransformers.ner import NERModel,NERArgs
```

```python
label=['obj','n-obj','a-obj','b-obj','c-obj','d-obj','e-obj','f-obj','g-obj']
```

```python
args = NERArgs()
args.num_train_epochs = 20
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32
```

```python
model = NERModel('bert', 'bert-base-cased',labels=label,use_cuda=False,args =args)
```

```python
from sklearn.metrics import accuracy_score
model.train_model(train_data,acc=accuracy_score)
```

```python
result, model_outputs, preds_list = model.eval_model(train_data)
result
```

```python
prediction, model_output = model.predict(["i have 2 cars"])
```
